```python
import matplotlib.pyplot as plt
import numpy as np
```

## Integrales

```python
def funcion_a_integrar(x):
    return x**2
a= 2
b= 5
N=10000
x= np.random.uniform(a,b,N)
y= np.random.uniform(0,funcion_a_integrar(b),N)
```

```python
n_bajo_curva= np.sum(y < funcion_a_integrar(x))
n_bajo_curva
```

```python
respuesta = (b-a)*funcion_a_integrar(b)*n_bajo_curva / N
respuesta
```

```python
plt.plot(x,y,'.')
plt.plot(np.linspace(a,b,100), funcion_a_integrar(np.linspace(a,b,100),),'r')
plt.show()
```

## CALCULAR PI (ÁREA DEL CIRCULO)

área del circulo, = PI* r^2
área del cuadrado = x^2 * y^2

```python
a= -1
b= 1
N=10000
x= np.random.uniform(a,b,N)
y= np.random.uniform(a,b,N)
circulo= x**2+y**2 <=a**2
n_dentro= np.sum(circulo)
pi = (n_dentro / N) * 4
pi
```

```python
plt.figure(figsize=(6,6))
plt.plot(x[circulo], y[circulo], 'b.', label='Dentro del círculo')
plt.plot(x[~circulo], y[~circulo], 'r.', label='Fuera del círculo')
plt.xlim(a, b)
plt.ylim(a, b)
plt.gca().set_aspect('equal', adjustable='box')
plt.title('Estimación de π')
plt.legend()
```
