### 1.Generar y graficar 1 caminata aleatoria

```python
import numpy as np
import matplotlib.pyplot as plt

np.random.seed(41)

n_steps =200
steps = np.random.choice([-1,1], size=n_steps)
price =100 + np.cumsum(steps)

plt.figure(figsize=(10,5))
plt.plot(price)
plt.title("Random Walk de un precio")
plt.xlabel("Tiempo")
plt.ylabel("Precio")
plt.show()
```

### 2. Generar y graficar la distribucin final de N caminatas aleatorias

```python
import numpy as np
import matplotlib.pyplot as plt

np.random.seed(1)

n_sim =50000
n_steps =200

final_positions = []

for _ in range(n_sim):
    steps = np.random.choice([-1,1], size=n_steps)
    final_positions.append(np.sum(steps))

plt.hist(final_positions, bins=50)
plt.title("Distribucin de posiciones finales")
plt.show()
```

### 3. Calcular la varianza de los resultados para diferentes largos de caminatas

```python
import numpy as np

np.random.seed(0)

n_sim = 2000
times = [10, 50, 100, 200, 400]

for t in times:
    finals = []
    for _ in range(n_sim):
        steps = np.random.choice([-1, 1], size=t)
        finals.append(np.sum(steps))
    print(f"t={t}, var  {np.var(finals):.2f}")
```

### 4.Generar y graficar una aproximacin de un Mov. Browniano con un random walk

```python
import numpy as np
import matplotlib.pyplot as plt

np.random.seed(42)

T =1
n =1000
dt = T/n

increments = np.random.normal(0, np.sqrt(dt), size=n)
W = np.cumsum(increments)

plt.plot(W)
plt.title("Aproximacin a movimiento browniano")
plt.show()

```
